```python
import PyPDF2

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    try:
        # Open the PDF file in binary read mode1
        with open('10554236.pdf', 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ""
            
            # Loop through each page in the PDF
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text()
            
            return text
    except Exception as e:
        print(f"Error: {str(e)}")
        return None

# Example usage:
pdf_path = '10554236.pdf'  # Replace with the path to our PDF file
pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text)

# Now we have the text content of the PDF in the 'pdf_text' variable
# we can implement logic to extract category, skills, and education from 'pdf_text'
```

```python
import re
# Extract Category (Job Role)
category_match = re.search(r'([A-Za-z\s]+)', pdf_text)
if category_match:
    category = category_match.group(1)
else:
    category = "N/A"

# Extract Skills using a specific regex pattern for semicolon-separated skills
skills_pattern = r'Skills\n(.*?)(?=\n\n|$)'
skills_match = re.search(skills_pattern, pdf_text, re.DOTALL)

if skills_match:
    skills_text = skills_match.group(1).strip()
    skills_list = [skill.strip() for skill in skills_text.split(';')]
else:
    skills_list = []

# Extract Education
education_match = re.search(r'Education\n(.*?)(?=\n\n|$)', pdf_text, re.DOTALL)
if education_match:
    education = education_match.group(1).strip().split('\n- ')
else:
    education = []

# Print the extracted details
print("Category:", category)
print("Skills:", skills_list)
print("Education:", education)
```

```python
import pandas as pd
import torch
from transformers import DistilBertTokenizer, DistilBertModel
from sklearn.metrics.pairwise import cosine_similarity

# Load your CSV data
dfr = pd.read_csv('Resume.csv')
cv_text_list = dfr['Resume_str'].tolist()[:3]

# Load the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased")

job_descriptions = df['job_description'].tolist()[:5]

# Tokenize and preprocess the job descriptions
job_description_tokens = tokenizer(job_descriptions, padding=True, truncation=True, return_tensors="pt")

# Define batch size for CVs (you can adjust this based on your available memory)
batch_size = 2
cv_embeddings = []

# Tokenize and preprocess CVs in batches
for i in range(0, len(cv_text_list), batch_size):
    batch_cv_texts = cv_text_list[i:i + batch_size]
    batch_cv_tokens = tokenizer(batch_cv_texts, padding=True, truncation=True, return_tensors="pt")
    
    # Calculate embeddings for the batch
    batch_cv_embeddings = model(**batch_cv_tokens).last_hidden_state
    
    cv_embeddings.append(batch_cv_embeddings)

# Concatenate the list of batch_cv_embeddings
cv_embeddings_concatenated = torch.cat(cv_embeddings, dim=0)

# Reshape cv_embeddings_concatenated to 2D
cv_embeddings_2d = cv_embeddings_concatenated.view(cv_embeddings_concatenated.shape[0], -1)

# Tokenize and preprocess job descriptions
job_description_tokens = tokenizer(job_descriptions, padding=True, truncation=True, return_tensors="pt")

# Calculate embeddings for job descriptions
job_description_embeddings = model(**job_description_tokens).last_hidden_state

# Reshape job_description_embeddings to 2D
job_description_embeddings_2d = job_description_embeddings.view(job_description_embeddings.shape[0], -1)

from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between job descriptions and CVs
similarity_matrix = cosine_similarity(job_description_embeddings_2d.detach().numpy(), cv_embeddings_2d.detach().numpy())

# Rank CVs based on similarity scores for each job description
ranked_candidates = []

for i, job_desc in enumerate(job_descriptions):
    similarity_scores = similarity_matrix[i]
    sorted_indices = similarity_scores.argsort()[::-1]  # Sort in descending order
    top_candidates = [(cv_text_list[idx], similarity_scores[idx]) for idx in sorted_indices]
    ranked_candidates.append((job_desc, top_candidates[:5]))  # Store the top 5 candidates for each job description
```
